```python
#Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()
def sum(x,y):
    return x+y
def my_reduce(func,seq):
    first=seq[0]
    for i in seq[1:]:
        first=func(first,i)
    return first
list=[1,2,3,4,5]
my_reduce(sum,list)
```

```python
#Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()
def even(n):
    if n%2==0:
        return True
    else:
        return False
def my_filter(func,seq):
    m=[]
    for i in seq:
        l=func(i)
        if l==True:
            m.append(i)
    print(m)
list=[1,2,3,4,5]
my_filter(even,list)
```

```python
#Implement a function longestWord() that takes a list of words and returns the longest one.
def longestword(a_list):
    n_len=[]
    for i in a_list:
        n_len.append((len(i),i))
    n_len.sort()
    return n_len[-1][1]
print(longestword(["arshiya","aman","saniya"]))
```

```python
"""Write a Python Program(with class concepts) to find the area of the triangle using the below formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass."""
class parent:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
class drived(parent):
    def __init__(self,*args):
        super(drived,self).__init__(*args)
    def area(self,*args):
        s=(self.a+self.b+self.c)/2
        area=(s*(s-self.a)*(s-self.b)*(s-self.c))
        return area
```

```python
y=drived(4,3,5)
```

```python
y.area()
```

```python
#Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.
def longwords(wordlist, length):
    return (word for word in wordlist if len(word) >= length)

def main():
    words = input("Enter words, separated by spaces: ").split()
    length = int(input("Minimum length of words to keep: "))
    print("Words longer than {} are {}.".format(length,"".join(longwords(words, length))))

main()
```

```python
#Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.
def vowel(char):
    if char=="a" or char=="e" or char=="i" or char=="o" or char=="u":
        return True
    else:
        return False 
```
